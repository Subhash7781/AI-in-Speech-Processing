### A1.

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

y, sr = librosa.load('Subbu.wav')

plt.figure(figsize=(10, 5))
librosa.display.waveshow(y, sr=sr, alpha=0.8,color='g')
plt.title('Speech Signal Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

### A2. 

```python
print("Length of the signal:",len(y),"samples")
print("Sampling rate:",sr, "Hz")
print("Magnitude range:",np.min(y),"to", np.max(y))
```

### A3. 

```python
import IPython.display as ipd

start = int(0 * sr)
end = int(3 * sr)
segment = y[start:end]

plt.figure(figsize=(10, 5))
librosa.display.waveshow(segment, sr=sr,color='b')
plt.title('Segment Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()

ipd.Audio(segment, rate=sr)
```

### A4.

```python
segment_1_start = int(3 * sr)
segment_1_end = int(5* sr)  
segment_2_start = int(6* sr)
segment_2_end = int(10* sr)
segment_1 = y[segment_1_start:segment_1_end]
segment_2 = y[segment_2_start:segment_2_end]
print("Playing Segment 1:")
plt.figure(figsize=(10, 5))
librosa.display.waveshow(segment_1, sr=sr,color='r')
plt.title('Segment-1 waveform')
plt.show()
ipd.Audio(segment_1, rate=sr)
```

```python
print("Playing Segment 2:")

plt.figure(figsize=(10, 5))
librosa.display.waveshow(segment_2)
plt.title('Segment-2 wave form')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()
ipd.Audio(segment_2, rate=sr)
```

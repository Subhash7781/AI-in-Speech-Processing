```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

# Load the audio file
file_path = "Subbu.wav"  # Replace with the path to your audio file
signal, sr = librosa.load(file_path, sr=None)

# Extract a portion of the signal (for example, the first 1 second)
signal_portion = signal[:sr]  # Assuming the sample rate is sr samples/second

# Perform FFT
fft_result = np.fft.fft(signal_portion)
magnitude_spectrum = np.abs(fft_result)

# Frequency axis
freq_axis = np.fft.fftfreq(len(signal_portion), d=1/sr)

# Plot amplitude spectrum
plt.figure(figsize=(10, 6))
plt.plot(freq_axis[:len(freq_axis)//2], magnitude_spectrum[:len(freq_axis)//2])
plt.title('Amplitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example signal snippets representing vowel sounds
# You'll need to replace these with your actual signal data
vowel_a = np.random.random(1000)  # Example snippet for 'a'
vowel_e = np.random.random(1000)  # Example snippet for 'e'
vowel_i = np.random.random(1000)  # Example snippet for 'i'

# Perform FFT
fft_a = np.fft.fft(vowel_a)
fft_e = np.fft.fft(vowel_e)
fft_i = np.fft.fft(vowel_i)

# Frequency axis
sampling_rate = 44100  # Example sampling rate, replace with your actual sampling rate
freq_axis = np.fft.fftfreq(len(vowel_a)) * sampling_rate

# Plot amplitude spectrum
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(freq_axis[:len(freq_axis)//2], np.abs(fft_a)[:len(freq_axis)//2])
plt.title('Amplitude Spectrum for Vowel "a"')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)
plt.plot(freq_axis[:len(freq_axis)//2], np.abs(fft_e)[:len(freq_axis)//2])
plt.title('Amplitude Spectrum for Vowel "e"')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 3)
plt.plot(freq_axis[:len(freq_axis)//2], np.abs(fft_i)[:len(freq_axis)//2])
plt.title('Amplitude Spectrum for Vowel "i"')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
